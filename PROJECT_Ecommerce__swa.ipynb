# Outlier

```python
# Loop over the columns in the DataFrame
for i in data.columns:
  # Check if the data type of the column is int64 or float64
  if((data[i].dtype == 'int64') | (data[i].dtype == 'float64')):
    # Set the title of the plot to the name of the column
    plt.title(i)
    # Create a box plot of the column
    sns.boxplot(data[i])
    # Show the plot
    plt.show()
```

```python
def iqr_capping(data,variables):
  q1=data[variables].quantile(0.25)
  q3=data[variables].quantile(0.75)
  iqr=q3-q1
  upper_whisker=q3+(1.5*iqr)
  lower_whisker=q1-(1.5*iqr)
  return lower_whisker,upper_whisker
```

```python
lower_lim,upper_lim=iqr_capping(data,'Tenure')
print('lower_lim =',lower_lim)
print('upper_lim =',upper_lim)
data['Tenure']=np.where(data['Tenure']>upper_lim,upper_lim,
                  np.where(data['Tenure']<lower_lim,lower_lim,data['Tenure']))
```

```python
lower_lim,upper_lim=iqr_capping(data,'WarehouseToHome')
print('lower_lim =',lower_lim)
print('upper_lim =',upper_lim)
data['WarehouseToHome']=np.where(data['WarehouseToHome']>upper_lim,upper_lim,
                  np.where(data['WarehouseToHome']<lower_lim,lower_lim,data['WarehouseToHome']))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
metric_k=[]
neighbors=np.arange(3,15)
```

```python
for k in neighbors:
    classifier=KNeighborsClassifier(n_neighbors=k,metric='minkowski',p=2)
    classifier.fit(x_train,y_train)
    y_pread=classifier.predict(x_test)
    acc=accuracy_score(y_test,y_pread)
    metric_k.append(acc)
```

```python
plt.plot(neighbors,metric_k,"o-")
plt.xlabel('k value')
plt.ylabel('accuracy')
plt.grid()
```
